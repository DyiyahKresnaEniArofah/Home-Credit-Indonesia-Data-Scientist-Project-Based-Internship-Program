##X. Comprehensive EDA Report

```python
def generate_eda_report(df_train):
    """
    EDA report dengan perbaikan minimal - hanya menambahkan kalkulasi target_correlation
    """

    # Hitung target correlation secara otomatis
    numerical_cols = df_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
    categorical_cols = df_train.select_dtypes(include=['object', 'category']).columns.tolist()

    # Inisialisasi target_correlation
    target_correlation = pd.Series(dtype=float)

    if 'TARGET' in df_train.columns and df_train['TARGET'].dtype in ['int64', 'float64']:
        correlation_matrix = df_train[numerical_cols].corr()
        if 'TARGET' in correlation_matrix.columns:
            target_correlation = correlation_matrix['TARGET'].drop('TARGET', errors='ignore')

    report = {
        'dataset_shape': df_train.shape,
        'missing_values_percentage': df_train.isnull().sum().sum() / (df_train.shape[0] * df_train.shape[1]) * 100,
        'numerical_features_count': len(numerical_cols),
        'categorical_features_count': len(categorical_cols),
        'target_distribution': df_train['TARGET'].value_counts(normalize=True).to_dict(),
        'high_correlation_features': target_correlation[abs(target_correlation) > 0.1].index.tolist() if not target_correlation.empty else []
    }

    print("=== EDA SUMMARY REPORT ===")
    print(f"Dataset Shape: {report['dataset_shape']}")
    print(f"Total Missing Values: {report['missing_values_percentage']:.2f}%")
    print(f"Numerical Features: {report['numerical_features_count']}")
    print(f"Categorical Features: {report['categorical_features_count']}")
    print(f"Target Distribution: {report['target_distribution']}")
    print(f"Features with |correlation| > 0.1 with target: {report['high_correlation_features']}")

    return report

eda_report = generate_eda_report(df_train)
```

##VI. Outlier Detection untuk Testing Data

```python
# Deteksi outliers di testing data - SIMPLE VERSION
def detect_test_outliers_simple(df_test):
    """
    Simple version dengan deteksi otomatis numerical features
    """
    # Deteksi numerical features otomatis
    numerical_features = df_test.select_dtypes(include=['int64', 'float64']).columns.tolist()
    print(f"Analyzing {len(numerical_features)} numerical features")

    outlier_report = {}

    for feature in numerical_features[:8]:  # Analisis 8 features pertama
        Q1 = df_test[feature].quantile(0.25)
        Q3 = df_test[feature].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df_test[(df_test[feature] < lower_bound) | (df_test[feature] > upper_bound)]
        outlier_percentage = len(outliers) / len(df_test) * 100

        outlier_report[feature] = {
            'outlier_count': len(outliers),
            'outlier_percentage': outlier_percentage,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        }

    return pd.DataFrame(outlier_report).T

# Deteksi outliers
outlier_analysis = detect_test_outliers_simple(df_test)
print("=== OUTLIER ANALYSIS ===")
print(outlier_analysis)

# Visualisasi outliers
plt.figure(figsize=(12, 8))
numerical_features = df_test.select_dtypes(include=['int64', 'float64']).columns.tolist()
outlier_features = numerical_features[:4]

for i, feature in enumerate(outlier_features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df_test, y=feature)
    plt.title(f'Outliers in {feature} - Testing Data')

plt.tight_layout()
plt.show()
```

##VII. Comprehensive Testing EDA Report

```python
def generate_testing_eda_report(train_df, test_df):
    report = {
        # Basic information
        'testing_shape': test_df.shape,
        'training_shape': train_df.shape,

        # Data quality
        'testing_missing_percentage': test_df.isnull().sum().sum() / (test_df.shape[0] * test_df.shape[1]) * 100,
        'training_missing_percentage': train_df.isnull().sum().sum() / (train_df.shape[0] * train_df.shape[1]) * 100,

        # Feature types
        'testing_numerical_features': len(test_df.select_dtypes(include=[np.number]).columns),
        'testing_categorical_features': len(test_df.select_dtypes(include=['object']).columns),

        # Data drift indicators
        'features_with_drift': [],
        'high_missing_features': [],
        'new_categorical_values': 0
    }

    # Check for distribution drift
    numerical_features = test_df.select_dtypes(include=[np.number]).columns
    for feature in numerical_features:
        if feature in train_df.columns:
            # Simple drift detection using KS test
            try:
                ks_stat, p_value = stats.ks_2samp(
                    train_df[feature].dropna(),
                    test_df[feature].dropna()
                )
                if p_value < 0.05:  # Significant difference
                    report['features_with_drift'].append(feature)
            except:
                pass

    # Check for high missing features
    missing_test = test_df.isnull().sum() / len(test_df)
    report['high_missing_features'] = missing_test[missing_test > 0.3].index.tolist()

    # Check for new categorical values
    categorical_features = test_df.select_dtypes(include=['object']).columns
    for feature in categorical_features:
        if feature in train_df.columns:
            train_categories = set(train_df[feature].dropna().unique())
            test_categories = set(test_df[feature].dropna().unique())
            report['new_categorical_values'] += len(test_categories - train_categories)

    return report

# Generate comprehensive report
eda_test_report = generate_testing_eda_report(df_train, df_test)

print("=== COMPREHENSIVE TESTING EDA REPORT ===")
print(f"Testing Dataset Shape: {eda_test_report['testing_shape']}")
print(f"Missing Values - Testing: {eda_test_report['testing_missing_percentage']:.2f}%")
print(f"Missing Values - Training: {eda_test_report['training_missing_percentage']:.2f}%")
print(f"Numerical Features in Testing: {eda_test_report['testing_numerical_features']}")
print(f"Categorical Features in Testing: {eda_test_report['testing_categorical_features']}")
print(f"Features with Distribution Drift: {len(eda_test_report['features_with_drift'])}")
print(f"Features with >30% Missing: {len(eda_test_report['high_missing_features'])}")
print(f"Total New Categorical Values: {eda_test_report['new_categorical_values']}")

# Recommendations based on EDA
print("\n=== RECOMMENDATIONS ===")
if len(eda_test_report['features_with_drift']) > 0:
    print("âš ï¸  WARNING: Distribution drift detected. Model performance may degrade.")
    print(f"   Affected features: {eda_test_report['features_with_drift'][:5]}")

if len(eda_test_report['high_missing_features']) > 0:
    print("âš ï¸  WARNING: High missing values in testing data.")
    print(f"   Features: {eda_test_report['high_missing_features']}")

if eda_test_report['new_categorical_values'] > 0:
    print("âš ï¸  WARNING: New categorical values in testing data.")
    print("   Consider how to handle unseen categories.")
```

##III. Relationship Analysis antara Kedua Dataset
Menganalisis bagaimana kedua dataset terhubung

Memahami pola pembayaran per aplikasi sebelumnya

Mengecek missing relationship antara aplikasi dan pembayaran

```python
# Analisis hubungan antara previous application dan installments
def analyze_relationship(prev_df, inst_df):
    # Group installments by SK_ID_PREV
    inst_summary = inst_df.groupby('SK_ID_PREV').agg({
        'AMT_INSTALMENT': ['count', 'sum'],
        'AMT_PAYMENT': ['sum'],
        'DAYS_ENTRY_PAYMENT': ['min', 'max']
    }).round(2)

    inst_summary.columns = ['NUM_INSTALMENTS', 'TOTAL_INSTALMENT', 'TOTAL_PAYMENT',
                           'FIRST_PAYMENT_DAY', 'LAST_PAYMENT_DAY']
    inst_summary = inst_summary.reset_index()

    # Merge dengan previous application
    merged_df = prev_df.merge(inst_summary, on='SK_ID_PREV', how='left', indicator=True)

    print("=== RELATIONSHIP ANALYSIS ===")
    print("Merge results:")
    print(merged_df['_merge'].value_counts())

    return merged_df, inst_summary

# Jalankan analisis hubungan
merged_data, inst_summary = analyze_relationship(prev_app, installments)
```

```python
# Analisis financial aspects previous application
def financial_analysis_prev(df):
    # Filter numerical financial columns
    financial_cols = ['AMT_APPLICATION', 'AMT_CREDIT', 'AMT_ANNUITY',
                     'AMT_GOODS_PRICE', 'AMT_DOWN_PAYMENT']

    financial_data = df[financial_cols].describe()
    print("=== FINANCIAL SUMMARY - PREVIOUS APPLICATIONS ===")
    print(financial_data)

    # Visualisasi distribusi amount
    fig, axes = plt.subplots(2, 3, figsize=(15, 10))
    axes = axes.ravel()

    for i, col in enumerate(financial_cols):
        if i < len(axes):
            # Filter outliers untuk visualisasi yang better
            data = df[col].dropna()
            Q1 = data.quantile(0.25)
            Q3 = data.quantile(0.75)
            IQR = Q3 - Q1
            filtered_data = data[(data >= Q1 - 1.5*IQR) & (data <= Q3 + 1.5*IQR)]

            axes[i].hist(filtered_data, bins=30, alpha=0.7, color='lightgreen')
            axes[i].set_title(f'Distribution of {col}')
            axes[i].set_xlabel('Amount')
            axes[i].set_ylabel('Frequency')

    # Hide unused subplots
    for i in range(len(financial_cols), len(axes)):
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

    # Analysis approval rates by loan characteristics
    if 'NAME_CONTRACT_STATUS' in df.columns:
        approval_analysis = df.groupby('NAME_CONTRACT_TYPE')['NAME_CONTRACT_STATUS'].value_counts(normalize=True).unstack()
        print("\n=== APPROVAL RATES BY CONTRACT TYPE ===")
        print(approval_analysis)

financial_analysis_prev(prev_app)
```

```python
# Analisis mendalam installments payments
def analyze_installments_payments(df):
    print("=== INSTALLMENTS PAYMENTS ANALYSIS ===")

    # Basic statistics
    print("Basic statistics of installments:")
    print(df[['AMT_INSTALMENT', 'AMT_PAYMENT']].describe())

    # Payment behavior analysis
    df['PAYMENT_DIFF'] = df['AMT_PAYMENT'] - df['AMT_INSTALMENT']
    df['PAYMENT_RATIO'] = df['AMT_PAYMENT'] / df['AMT_INSTALMENT']
    df['DAYS_DIFF'] = df['DAYS_ENTRY_PAYMENT'] - df['DAYS_INSTALMENT']

    print("\n=== PAYMENT BEHAVIOR ===")
    print(f"Early payments: {len(df[df['DAYS_DIFF'] < 0])} ({len(df[df['DAYS_DIFF'] < 0])/len(df)*100:.2f}%)")
    print(f"On-time payments: {len(df[df['DAYS_DIFF'] == 0])} ({len(df[df['DAYS_DIFF'] == 0])/len(df)*100:.2f}%)")
    print(f"Late payments: {len(df[df['DAYS_DIFF'] > 0])} ({len(df[df['DAYS_DIFF'] > 0])/len(df)*100:.2f}%)")

    # Payment amount behavior
    print(f"\nUnder payments: {len(df[df['PAYMENT_RATIO'] < 1])} ({len(df[df['PAYMENT_RATIO'] < 1])/len(df)*100:.2f}%)")
    print(f"Full payments: {len(df[df['PAYMENT_RATIO'] == 1])} ({len(df[df['PAYMENT_RATIO'] == 1])/len(df)*100:.2f}%)")
    print(f"Over payments: {len(df[df['PAYMENT_RATIO'] > 1])} ({len(df[df['PAYMENT_RATIO'] > 1])/len(df)*100:.2f}%)")

    return df

installments = analyze_installments_payments(installments)
```

```python
# Visualisasi perilaku pembayaran
def visualize_payment_behavior(df):
    fig, axes = plt.subplots(2, 2, figsize=(15, 12))

    # 1. Distribution of payment delays
    axes[0,0].hist(df['DAYS_DIFF'].dropna(), bins=50, alpha=0.7, color='lightcoral')
    axes[0,0].axvline(x=0, color='red', linestyle='--', label='Due Date')
    axes[0,0].set_title('Distribution of Payment Delays (Days)')
    axes[0,0].set_xlabel('Days Difference (Entry - Instalment)')
    axes[0,0].set_ylabel('Frequency')
    axes[0,0].legend()

    # 2. Payment ratio distribution
    payment_ratio = df['PAYMENT_RATIO'].dropna()
    payment_ratio = payment_ratio[payment_ratio.between(0, 2)]  # Filter outliers
    axes[0,1].hist(payment_ratio, bins=50, alpha=0.7, color='lightgreen')
    axes[0,1].axvline(x=1, color='red', linestyle='--', label='Full Payment')
    axes[0,1].set_title('Distribution of Payment Ratios')
    axes[0,1].set_xlabel('Payment Ratio (Actual/Expected)')
    axes[0,1].set_ylabel('Frequency')
    axes[0,1].legend()

    # 3. Payment timing vs amount
    scatter_data = df.sample(1000)  # Sample for performance
    axes[1,0].scatter(scatter_data['DAYS_DIFF'], scatter_data['PAYMENT_RATIO'],
                     alpha=0.6, color='blue')
    axes[1,0].axvline(x=0, color='red', linestyle='--')
    axes[1,0].axhline(y=1, color='red', linestyle='--')
    axes[1,0].set_title('Payment Timing vs Amount Ratio')
    axes[1,0].set_xlabel('Days Difference')
    axes[1,0].set_ylabel('Payment Ratio')

    # 4. Monthly payment pattern
    df['MONTH'] = (df['DAYS_INSTALMENT'] / 30).astype(int)
    monthly_pattern = df.groupby('MONTH').agg({
        'PAYMENT_RATIO': 'mean',
        'DAYS_DIFF': 'mean'
    }).reset_index()

    axes[1,1].plot(monthly_pattern['MONTH'], monthly_pattern['PAYMENT_RATIO'],
                  marker='o', label='Payment Ratio')
    axes[1,1].set_title('Payment Behavior Over Time')
    axes[1,1].set_xlabel('Month')
    axes[1,1].set_ylabel('Average Payment Ratio')
    axes[1,1].legend()

    plt.tight_layout()
    plt.show()

visualize_payment_behavior(installments)
```

```python
# Analisis gabungan - riwayat pembayaran per aplikasi
def merged_analysis(prev_df, inst_df):
    # Aggregasi payment behavior per SK_ID_PREV
    payment_behavior = inst_df.groupby('SK_ID_PREV').agg({
        'AMT_INSTALMENT': ['count', 'sum'],
        'AMT_PAYMENT': 'sum',
        'DAYS_DIFF': ['mean', 'std', 'max'],
        'PAYMENT_RATIO': ['mean', 'min', 'max']
    }).round(4)

    # Flatten column names
    payment_behavior.columns = [
        'NUM_PAYMENTS', 'TOTAL_INSTALMENT', 'TOTAL_PAYMENT',
        'AVG_DAYS_DIFF', 'STD_DAYS_DIFF', 'MAX_DAYS_DELAY',
        'AVG_PAYMENT_RATIO', 'MIN_PAYMENT_RATIO', 'MAX_PAYMENT_RATIO'
    ]

    payment_behavior = payment_behavior.reset_index()

    # Merge dengan previous application
    merged = prev_df.merge(payment_behavior, on='SK_ID_PREV', how='left')

    # Analysis by previous application status
    if 'NAME_CONTRACT_STATUS' in merged.columns:
        status_analysis = merged.groupby('NAME_CONTRACT_STATUS').agg({
            'AVG_DAYS_DIFF': 'mean',
            'AVG_PAYMENT_RATIO': 'mean',
            'MAX_DAYS_DELAY': 'mean',
            'NUM_PAYMENTS': 'mean'
        }).round(2)

        print("=== PAYMENT BEHAVIOR BY PREVIOUS APPLICATION STATUS ===")
        print(status_analysis)

    return merged

merged_complete = merged_analysis(prev_app, installments)
```

### A. AGGREGATE FEATURES FROM PREVIOUS_APPLICATION

```python
def aggregate_previous_applications(prev_df):
    """Aggregate previous application data by SK_ID_CURR"""

    # Basic aggregation
    prev_agg = prev_df.groupby('SK_ID_CURR').agg({
        'AMT_ANNUITY': ['mean', 'max', 'min', 'sum'],
        'AMT_APPLICATION': ['mean', 'max', 'min', 'sum'],
        'AMT_CREDIT': ['mean', 'max', 'min', 'sum'],
        'AMT_DOWN_PAYMENT': ['mean', 'max', 'min', 'sum'],
        'AMT_GOODS_PRICE': ['mean', 'max', 'min', 'sum'],
        'HOUR_APPR_PROCESS_START': ['mean', 'max', 'min'],
        'RATE_DOWN_PAYMENT': ['mean', 'max', 'min'],
        'DAYS_DECISION': ['mean', 'max', 'min'],
        'SELLERPLACE_AREA': ['mean', 'max', 'min'],
        'CNT_PAYMENT': ['mean', 'sum'],

        # Flag for previous applications behavior
        'NAME_CONTRACT_TYPE': ['nunique', 'count'],
        'WEEKDAY_APPR_PROCESS_START': ['nunique'],
        'FLAG_LAST_APPL_PER_CONTRACT': ['mean'],
        'NFLAG_LAST_APPL_IN_DAY': ['mean'],

        # Application status flags
        'NAME_CONTRACT_STATUS': lambda x: x.value_counts().to_dict()
    }).reset_index()

    # Flatten column names
    prev_agg.columns = ['PREV_' + '_'.join(col).strip() if col[0] != 'SK_ID_CURR'
                       else 'SK_ID_CURR' for col in prev_agg.columns.values]

    # Create specific flags from contract status
    status_agg = prev_df.groupby('SK_ID_CURR')['NAME_CONTRACT_STATUS'].value_counts().unstack(fill_value=0)
    status_agg.columns = ['PREV_STATUS_' + str(col) for col in status_agg.columns]
    status_agg = status_agg.reset_index()

    # Merge with main aggregation
    prev_agg = prev_agg.merge(status_agg, on='SK_ID_CURR', how='left')

    return prev_agg
```

### B. AGGREGATE FEATURES FROM INSTALLMENTS_PAYMENTS

```python
def aggregate_installments_payments(inst_df):
    """Aggregate installments payments data by SK_ID_CURR"""

    # Calculate payment behavior metrics
    inst_df['PAYMENT_DIFF'] = inst_df['AMT_INSTALMENT'] - inst_df['AMT_PAYMENT']
    inst_df['PAYMENT_RATIO'] = inst_df['AMT_PAYMENT'] / inst_df['AMT_INSTALMENT']
    inst_df['DAYS_PAYMENT_EARLY'] = inst_df['DAYS_INSTALMENT'] - inst_df['DAYS_ENTRY_PAYMENT']

    inst_agg = inst_df.groupby('SK_ID_CURR').agg({
        'AMT_INSTALMENT': ['mean', 'max', 'min', 'sum'],
        'AMT_PAYMENT': ['mean', 'max', 'min', 'sum'],
        'PAYMENT_DIFF': ['mean', 'max', 'min', 'sum'],
        'PAYMENT_RATIO': ['mean', 'max', 'min'],
        'DAYS_PAYMENT_EARLY': ['mean', 'max', 'min'],
        'NUM_INSTALMENT_VERSION': ['nunique'],
        'NUM_INSTALMENT_NUMBER': ['max']
    }).reset_index()

    # Flatten column names
    inst_agg.columns = ['INST_' + '_'.join(col).strip() if col[0] != 'SK_ID_CURR'
                       else 'SK_ID_CURR' for col in inst_agg.columns.values]

    # Late payment flags
    late_payments = inst_df[inst_df['DAYS_ENTRY_PAYMENT'] > inst_df['DAYS_INSTALMENT']]
    if not late_payments.empty:
        late_agg = late_payments.groupby('SK_ID_CURR').agg({
            'SK_ID_PREV': 'count',
            'DAYS_ENTRY_PAYMENT': 'mean'
        }).reset_index()
        late_agg.columns = ['SK_ID_CURR', 'INST_LATE_PAYMENT_COUNT', 'INST_AVG_LATE_DAYS']

        inst_agg = inst_agg.merge(late_agg, on='SK_ID_CURR', how='left')
        inst_agg['INST_LATE_PAYMENT_COUNT'] = inst_agg['INST_LATE_PAYMENT_COUNT'].fillna(0)
        inst_agg['INST_AVG_LATE_DAYS'] = inst_agg['INST_AVG_LATE_DAYS'].fillna(0)

    return inst_agg
```

###C. MERGE ALL DATASETS

```python
def create_enhanced_dataset(main_df, prev_df, inst_df):
    """Create enhanced dataset by merging all sources"""

    print("Original main dataset shape:", main_df.shape)

    # Aggregate features
    prev_agg = aggregate_previous_applications(prev_df)
    inst_agg = aggregate_installments_payments(inst_df)

    print("Previous applications aggregated shape:", prev_agg.shape)
    print("Installments payments aggregated shape:", inst_agg.shape)

    # Merge with main dataset
    enhanced_df = main_df.merge(prev_agg, on='SK_ID_CURR', how='left')
    enhanced_df = enhanced_df.merge(inst_agg, on='SK_ID_CURR', how='left')

    print("Enhanced dataset shape:", enhanced_df.shape)

    # Fill NaN values for new features
    prev_cols = [col for col in enhanced_df.columns if col.startswith('PREV_')]
    inst_cols = [col for col in enhanced_df.columns if col.startswith('INST_')]

    enhanced_df[prev_cols] = enhanced_df[prev_cols].fillna(0)
    enhanced_df[inst_cols] = enhanced_df[inst_cols].fillna(0)

    return enhanced_df
```

###D. APPLY TO TRAINING AND TEST DATA

```python
def create_enhanced_dataset_selective(main_df, previous_app_df, installments_df):
    """
    Enhanced dataset dengan selective aggregation berdasarkan data types
    """
    print("Creating enhanced features with selective aggregation...")

    # 1. Pre-process: Identifikasi numerical vs categorical columns
    def get_numerical_columns(df):
        return df.select_dtypes(include=['int64', 'float64']).columns.tolist()

    def get_categorical_columns(df):
        return df.select_dtypes(include=['object', 'category']).columns.tolist()

    # 2. Previous applications aggregation
    print("Processing previous applications...")
    prev_numerical = get_numerical_columns(previous_app_df)
    prev_categorical = get_categorical_columns(previous_app_df)

    # Hanya gunakan kolom numerical untuk aggregation statistical
    prev_agg_dict = {}

    # Numerical aggregations
    for col in prev_numerical:
        if col != 'SK_ID_CURR':  # Skip ID column
            prev_agg_dict[col] = ['mean', 'min', 'max', 'sum']

    # Categorical aggregations (hitung count dan unique values)
    for col in prev_categorical:
        prev_agg_dict[col] = ['count', 'nunique']

    prev_agg = previous_app_df.groupby('SK_ID_CURR').agg(prev_agg_dict).round(2)
    prev_agg.columns = ['PREV_' + '_'.join(col).upper() for col in prev_agg.columns]
    prev_agg = prev_agg.reset_index()

    # 3. Installments payments aggregation
    print("Processing installments payments...")
    inst_numerical = get_numerical_columns(installments_df)

    inst_agg_dict = {}
    for col in inst_numerical:
        if col != 'SK_ID_CURR':
            inst_agg_dict[col] = ['mean', 'min', 'max', 'sum']

    inst_agg = installments_df.groupby('SK_ID_CURR').agg(inst_agg_dict).round(2)
    inst_agg.columns = ['INST_' + '_'.join(col).upper() for col in inst_agg.columns]
    inst_agg = inst_agg.reset_index()

    # 4. Merge datasets
    print("Merging datasets...")
    enhanced_df = main_df.copy()
    enhanced_df = enhanced_df.merge(prev_agg, on='SK_ID_CURR', how='left')
    enhanced_df = enhanced_df.merge(inst_agg, on='SK_ID_CURR', how='left')

    # 5. Fill missing values
    new_cols = [col for col in enhanced_df.columns if col.startswith(('PREV_', 'INST_'))]
    enhanced_df[new_cols] = enhanced_df[new_cols].fillna(0)

    print(f"Enhanced dataset created: {enhanced_df.shape}")
    return enhanced_df

# Gunakan versi selective
print("Processing training data...")
df_train_enhanced = create_enhanced_dataset_selective(df_train, previous_app, installments_payments)

print("\nProcessing test data...")
df_test_enhanced = create_enhanced_dataset_selective(df_test, previous_app, installments_payments)
```
